```python
import pandas as pd
import json
import requests
import time
from multiprocessing import Pool
```

#### Import files

```python
# API keys
api_key = open('/Users/nzhang/data_science_projects/recommender_system/apikey.txt','r').read()
```

```python
# 5000 User IDs
file1 = open('/Users/nzhang/data_science_projects/recommender_system/steam_user_id.txt')
user_ids = [line.rstrip("\n") for line in file1.readlines() if line.strip()]
```

```python
# 500 App IDs
file2 = open('/Users/nzhang/data_science_projects/recommender_system/app_id.txt')
app_ids = [line.rstrip("\n") for line in file2.readlines() if line.strip()]
```

#### *Retrieve and store data of 5000 users*

```python
%%time

user_data = []

for user_steam_id in user_ids:    
    try:
        base_url = 'https://api.steampowered.com/IPlayerService/GetOwnedGames/v0001/'
        params = {
            'key': api_key,
            'steamid': user_steam_id,
            'format':'json'
        }
        user_info = requests.get(base_url, params=params).json().get('response').get('games')
        if len(user_info) > 0:
            user_game_inventory = user_info['games']
        else:
            user_game_inventory = []
        user_data.append({user_steam_id : user_game_inventory})
        
    except:
        print(user_steam_id)
```

```python
p = Pool(2)
```

```python
len(user_data)
```

```python
# Write users json data to text file
with open('user_game_data.txt', 'w') as outfile:
    for hostDict in user_data:
        json.dump(hostDict, outfile)
        outfile.write('\n')
```

#### *Retrieve and store data of 500 games*
Please keep in mind that you can only call this API 200 times every 5 minutes

https://steamspy.com/api.php

```python
%%time

app_data = []

for app_id in app_ids:
    try:
        app_url = 'http://steamspy.com/api.php?request=appdetails&appid=' + app_id
        app_detail = requests.get(app_url).json()
        app_data.append(app_detail)
        
    except:
        print(app_data)
        
    time.sleep(1.5)
```

```python
len(app_data)
```

```python
# Write games json data to text file
with open('game_detail_data.txt', 'w') as outfile:
    for hostDict in app_data:
        json.dump(hostDict, outfile)
        outfile.write('\n')
```
